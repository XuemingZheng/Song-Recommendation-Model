```python
# Question 2
# Here we define the user engagement is measured through the average plays per user in a given state.
state_play_count = data.groupby('user_state')['id'].count().reset_index()
state_play_count = state_play_count.rename(columns = {'id':'play_count'})
state_user_play = pd.merge(left = state_user_count,right = state_play_count,on='user_state')
state_user_play['average_play'] = state_user_play['play_count']/state_user_play['user_count']

state_user_play = state_user_play.sort_values(by='average_play',ascending = False)
```

```python
def find_first_user(df):
    """function to find the first user"""
    idx = df['user_sign_up_date'].argmin()
    return df.loc[idx,['user_id','user_sign_up_date']]
```

```python
data.groupby('user_state').apply(find_first_user).sort_values(by='user_sign_up_date')
```

```python
# here, Iimplement a simple ersion of collaborative filtering algorithm for song recommendation
song_user = data.groupby(['song_played','user_id'])['id'].count().unstack(fill_value=0)
song_user = (song_user > 0).astype(int)
song_user.head()
```

```python
song_user_norm = normalize(song_user, axis = 1)
similarity = np.dot(song_user_norm, song_user_norm.T)
similarity_df = pd.DataFrame(similarity, index = song_user.index, columns=song_user.index)

similarity_df.head()
```

```python
def find_topk(song, similarity,k=1):
    df = similarity.loc[song].sort_values(ascending=False)[1:k + 1].reset_index()
    df = df.rename(columns ={'song_played':'song',song:'Similarity'})
    
    return df
```

```python
df = find_topk(song='A Day In The Life',similarity=similarity_df,k=10)
df
```

## Design A/B test for song recommendation

Step 1: Define Goals and form hypothesis.
Goal: Identify the impact of the new song recommendation on user behavior.
Hypothesis: The song recommendation system will inspire user to spend more time in listening music.

Step 2: Identify treatment and control
Control: User who login and listen to the song with random recommendation. (previous set up)
Treatment: User who login and listen to the song with song recommendation system.

Step 3: Identify key metrics to measure
•	Evaluation metric: The average number of music play per day for active user (play at least one song at that day)
•	Invariant metric: The number of user login to listen to music.

Step 4: Identify data to be collected
•	The unit of diversion is user_id
•	User login time
•	What platform the user is on
•	Music loads time

Step 5: Make sure logging is in place
Dose it works on mobile? How does it work when user is logged out? (make sure when the user logout and re login to his/her account, he/she still see the same layout).

Step 6: Determine how small of a difference needs to detect.
Suppose the current average number of music play per day for active user is 3.
We set the minimum difference to be 10%.
Then we expected for the test group, the average number of music play per day for active user should at least increase to 3.3.

Step 7: Determine what fraction of visitors you want to be in the treatment.
The easiest way is to randomly divides the user into two equal parts. 
However, sometimes we want to be conservative, only assign a smaller portion of the user to test group.

Step 8: Calculate how much data to collect and how long you need to run the experiment for.
•	Minimum measurable difference (10% in our case)
•	Type 1 error (false positive) usually set to 0.05
•	Type 2 error (1-power) (False negative) usually set to 0.2
•	Expected standard deviation of the change in the metric.

Step 9: Figure out how long to run the test and run test for at least that long
Based on how many users log in to listen to music every day.
